```python
ID_list = pd.read_csv("test.csv")['ID']
predict_list = [predict_rf]
name = ['rf']

print(predict_list)
```

```python
def submission_csv(predict_list, name):
    for i in range(len(predict_list)):
        submission = pd.DataFrame({
            "ID":ID_list,
            "class":predict_list[i]['Label']
        })
        
        submission.to_csv('submission_'+name[i]+'.csv', index= False)

submission_csv(predict_list,name)
```
